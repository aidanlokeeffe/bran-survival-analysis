The purpose of this notebook is to test the brancher-annihilator class stored in bran.py. Below, we carry out a small simulation.

```python
import numpy as np
import matplotlib.pyplot as plt

from bran import BrAn
```

```python
%matplotlib widget
```

```python
n = 5
A = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        A[i,j] = np.random.choice([0,1])
print(A)
```

```python
p = 0.05
Alpha = 10000
```

```python
bran_instance = BrAn(A, p, Alpha)
```

```python
survival_times = bran_instance.simulate()
```

This is good. Yes, from here, we get a dictionary of survival times. I want to look at a histogram of these survival times.

```python
survival_array = np.array( [survival_times[alpha] for alpha in survival_times.keys()] )
```

```python
max_survival = np.max(survival_array)
print(max_survival)
```

```python
plt.figure()
plt.hist(survival_array, bins=max_survival)
plt.xlabel("Survival Time")
plt.ylabel("Count")
plt.title("Survival Time Distribution")
plt.show()
```

This time, we got mean 2.3901 and standard deviation 1.0722508988105348. Right skewed with max = 9.

Obviously, this is a small sample, but it seems that the result is the same. A more precise estimate may be in order.

Let's do n=1000. This also allows us to estimate the time complexity, although we're definitely not getting the worst case here.

```python
n = 1000
A = np.zeros((n,n))
for i in range(n):
    for j in range(n):
        A[i,j] = np.random.choice([0,1])

p = 0.05
Alpha = 10000

bran_instance = BrAn(A, p, Alpha)
survival_times = bran_instance.simulate()
survival_array = np.array( [survival_times[alpha] for alpha in survival_times.keys()] )
max_survival = np.max(survival_array)
print(max_survival)

plt.figure()
plt.hist(survival_array, bins=max_survival)
plt.xlabel("Survival Time")
plt.ylabel("Count")
plt.title("Survival Time Distribution")
plt.show()

print(np.mean(survival_array))
print(np.std(survival_array))
```

For a graph with 1000 nodes, the mean survival time was 1.9999 and the standard deviation was 0.00999949998749938. This time, the max was 2. Since the support of this distribution is so small, we can list the probability masses out.

```python
probabilities = {1:0, 2:0}
for alpha in survival_times.keys():
    probabilities[ survival_times[alpha] ] += 1
print(probabilities)
```

Wow! So of the 10000 messages, 9999 survived 2 time steps, and only 1 survived one time step. That's a 99% chance of surviving exactly 2 time steps. 

It really seems like, for an ER(1/2) graph, the expected survival time converges to 2 as $n \rightarrow \infty.$ In fact, it seems like the survival time itself converges to 2, no expectation necessary.
